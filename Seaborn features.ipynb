## Barplot - гистограмма

Классический график, хорошо применим к категориальным переменным

```python
titanic_dataset = sns.load_dataset("titanic")

sns.barplot(x="sex", y="survived", data=titanic_dataset)
```

Можно перевернуть

```python
sns.barplot(y="sex", x="survived", data=titanic_dataset)
```

Цвета графика можно изменять с помощью аргументов "color" и "palette"

```python
sns.barplot(x="sex", y="survived", data=titanic_dataset, color='blue')
```

```python
sns.barplot(x="embark_town", 
            y="survived", 
            palette='hls', 
            data=titanic_dataset)
```

Полезная функция - группировка данных на графике. Используется для сравнения значений по категориям и прочего. Для применения используем аргумент "hue"

```python
sns.barplot(x="class", y="survived", hue="embark_town", data=titanic_dataset)
```

Можно изменить порядок следования столбцов по умолчанию. Это делается с помощью аргумента "order"

```python
sns.barplot(x="class",
            y="survived",
            hue="embark_town",
            order=["Third", "Second", "First"],
            data=titanic_dataset)
```

Выдаваемый доверительный интервал на графике можно изменить или убрать вообще. Чтобы убрать, необхожимо установить значение аргумента "ci" = "None". Можно использовать стандартное отклонение вместо среднего, установив "ci"="sd". Можно также установить длину штриха на шкале ошибок, установив "capsize" = нужному ограничению

```python
sns.barplot(x="class",
            y="survived",
            hue="embark_town",
            ci=None,
            data=titanic_dataset)
```

```python
sns.barplot(x="class",
            y="survived",
            hue="embark_town",
            ci='sd',
            data=titanic_dataset)
```

```python
sns.barplot(x="class",
            y="survived",
            hue="embark_town",
            ci='sd',
            capsize=0.1,
            data=titanic_dataset)
```

## Heatmap

```python
sns.heatmap(titanic_dataset.corr())
```

Можно покапаться в аргументах<br />
<br />
<br />
vmin, vmax — устанавливают диапазон значений, которые служат основой для цветовой карты (colormap).<br />
<br />
cmap — определяет конкретную colormap, которую мы хотим использовать (ознакомьтесь с полным диапазоном цветовых палитр здесь).<br />
<br />
center — принимает вещественное число для центрирования цветовой карты; если cmap не указан, используется colormap по умолчанию; если установлено значение True — все цвета заменяются на синий.<br />
<br />
annot — при значении True числовые значения корреляции отображаются внутри ячеек.<br />
<br />
cbar — если установлено значение False, цветовая шкала (служит легендой) исчезает.<br />
<br />
fmt - String formatting code to use when adding annotations.<br />
<br />
linewidths - ширина линий, которые будут разделять каждую ячейку.<br />
<br />
linecolor - цвета линий, которые будут разделять каждую ячейку.<br />
<br />
xticklabels и yticklabels - включение/отключение x-label и y-label, по умолчанию они есть (=True)

```python
sns.heatmap(titanic_dataset.corr(), vmin=0.5)
```

```python
sns.heatmap(titanic_dataset.corr(), cmap="YlGnBu")
plt.show()

sns.heatmap(titanic_dataset.corr(), cmap="Blues")
plt.show()

sns.heatmap(titanic_dataset.corr(), cmap="BuPu")
plt.show()

sns.heatmap(titanic_dataset.corr(), cmap="Greens")
plt.show()

sns.heatmap(titanic_dataset.corr(), cmap='BrBG') #очень кайфовый
plt.show()
```

```python
heatmap = sns.heatmap(titanic_dataset.corr(),
                      vmin=-1,
                      vmax=1,
                      annot=True,
                      cmap='BrBG',
                      linewidths=4,
                      linecolor='black'
                      )
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':30}, pad=10)
#работа с заголовком: 1 - название, 2 - размер, 3 -удаленность от графика
```

## Треугольная heatmap

```python
# Определите маску, чтобы установить значения в верхнем треугольнике на True 
mask = np.triu(np.ones_like(titanic_dataset.corr(), dtype=np.bool)) 
heatmap = sns.heatmap(titanic_dataset.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG') 
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16)
```

## Корреляция независимых переменных с зависимой

```python
heatmap = sns.heatmap(titanic_dataset.corr()[['survived']].sort_values(by='survived', ascending=False),
            vmin=-1,
            vmax=1,
            annot=True,
            cmap='BrBG',
            cbar=False)
heatmap.set_title('Features Correlating with "Survived"', fontdict={'fontsize':18}, pad=16)
```

Если необходимо разделить общий box plot на несколько для отдельных признаков, то это можно сделать. Определите область отрисовки (fig) и нужное количество координатных осей (axes) с помощью функции subplots из Matplotlib. Доступ к нужной области объекта axes можно получить через его индекс. Функция boxplot() принимает ax аргумент, который по индексу объекта axes получает область для построения диаграммы

```python
fig, axes = plt.subplots(1, 2)
sns.boxplot(x=df['day'], y=DMC, orient='v', ax=axes[0])
sns.boxplot(x=df['day'], y=DC, orient='v', ax=axes[1])
```

## Boxplot с диаграммой рассеивания

Для более наглядного восприятия распределения можно наложить точечную диаграмму рассеивания на блочную

```python
df_1 = pd.DataFrame(data=df, columns=["DC", "DMC"])
boxplot = sns.boxplot(x="variable", y="value", data=pd.melt(df_1), order=["DC", "DMC"])
boxplot = sns.stripplot(x="variable",
                        y="value",
                        data=pd.melt(df_1),
                        marker="o", #чем обозначаются элементы рассеивания (символ)
                        alpha=0.5, #прозрачность цвета рассеивания (0 - прозрачный, 1 - обычный, 0.3 - средне)
                        color="green", #цвет рассеивания
                        order=["DC", "DMC"])
boxplot.axes.set_title("Распределение показателей при лесном пожаре", fontsize=16)
boxplot.set_xlabel("Показатели", fontsize=14)
boxplot.set_ylabel("Значения", fontsize=14)
```

## Violin Plot

Violin Plot или скрипичные диаграммы используются для визуализации распределения данных, отображая диапазон данных, медиану и область распределения данных

```python
sns.violinplot(x=life_exp) # или по классике sns.violinplot(x="life_exp", data=df)
```

## Violin Plot с диаграммой рассеивания

Точечную диаграмму распределения можно наложить на скрипичную диаграмму, чтобы увидеть размещение точек, составляющих это распределение. Для этого просто создается одна область рисования, а затем последовательно в ней создаются две диаграммы

```python
colors_list = [
    '#78C850', '#F08030', '#6890F0', 
    '#A8B820', '#F8D030', '#E0C068', 
    '#C03028', '#F85888', '#98D8D8'
]
plt.figure(figsize=(16,8))
sns.violinplot(x=continent, y=life_exp,palette=colors_list)
sns.swarmplot(x=continent, y=life_exp, color="k", alpha=0.5) #альфа также отвечает за прозрачность цвета
plt.title("Ожидаемая продолжительность жизни по континентам")
plt.ylabel("Ожидаемая продолжительность жизни")
plt.xlabel("Континент")
```

Если нужно разделить визуализацию столбцов из набора данных на их собственные диаграммы, то это можно сделать. Создайте область рисования и сетку, в ячейках которой будут графики.

Для отображения диаграммы в соответствующей ячейке применяется функция add_subplot(), которой передается адрес этой ячейки. Создание диаграммы делаем, как обычно, используя объект axes.

Можно использовать y=variable, либо data=variable

```python
fig = plt.figure(figsize=(6, 4))
gs = fig.add_gridspec(1, 3)

ax = fig.add_subplot(gs[0, 0])

sns.violinplot(data=population)
ax.set_xlabel("Население")

ax = fig.add_subplot(gs[0, 1])
sns.violinplot(data=life_exp)
ax.set_xlabel("Прод. жизни")

ax = fig.add_subplot(gs[0, 2])
sns.violinplot(data=gdp_cap)
ax.set_xlabel("Объем ВВП")

fig.tight_layout()
```

## Группировка скрипичных диаграмм по категориальному признаку

Например, если есть категориальная величина, имеющая два значения (обычно, True/False), то в этом случае можно группировать графики по этим значениям.

Допустим, есть набор данных по трудоустройству населения со столбцом employment и его значениями employed и unemployed. Тогда можно сгруппировать диаграммы по видам занятости

```python
europe = df.loc[df["continent"] == "Europe"]

# Вычисляем среднее значение переменной "life_exp"
avg_life_exp = df["life_exp"].mean()

# Добавим новую колонку
europe.loc[:, "above_average_life_exp"] = europe["life_exp"] > avg_life_exp
europe["above_average_life_exp"].replace(
    {True: "Yes", False: "No"}, 
    inplace=True
)
```

```python
europe = europe.head(70)
plt.figure(figsize=(16,8))
ax = sns.violinplot(x=europe.country, y=europe.life_exp, hue=europe.above_average_life_exp)
ax.set_title("Ожидаемая продолжительность жизни по странам")
ax.set_ylabel("Ожидаемая продолжительность жизни")
ax.set_xlabel("Страны")
```

```python
europe = europe.tail(50)

ax = sns.violinplot(
    x=europe.country, 
    y=europe.life_exp, 
    hue=europe.above_average_life_exp,
    split=True
)
ax.set_title("Ожидаемая продолжительность жизни по странам")
ax.set_ylabel("Ожидаемая продолжительность жизни")
ax.set_xlabel("Страны")
```
